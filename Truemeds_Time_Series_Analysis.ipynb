# Missing Value Detection

```python
df.isna().sum()
```

```python
Null_value=df.isna().sum().reset_index()
Null_value['% of Null Value']= Null_value[0]/df.shape[0]*100
Null_valuee=Null_value.sort_values(by='% of Null Value', ascending=False)
```

```python
plt.figure(figsize=(20, 3))
sns.barplot(data=Null_valuee, x='index', y='% of Null Value', color='blue')
plt.title('% of Null Value in Dataset')
plt.xticks(rotation=90)
plt.xlabel('Columns')
plt.ylabel('% of Null Value')
plt.show()
```

```python
#calculating mean value of Event revenue column to fill the missing value
mean_rev=df['Event Revenue'].mean()
df['Event Revenue'].fillna(mean_rev, inplace=True)
df['no_of_item'].fillna(0, inplace=True)
df['packaging_charge_amount'].fillna(0, inplace=True)
df['Carrier'].fillna('Unknown', inplace=True)
df['Operator'].fillna('Unknown', inplace=True)
df['customer_id'].fillna('Unknown', inplace=True)
df['has_coupon_code'].fillna("NA", inplace=True)
df['coupon_applied'].fillna("NA", inplace=True)
df['discount_amount'].fillna(0, inplace=True)
df['delivery_charge_amount'].fillna(0, inplace=True)
df['savings_amount'].fillna(0, inplace=True)
df['coupon_discount_amount'].fillna(0, inplace=True)
```

# Analysing Delta between Install Time and Event Time

```python
# calculating time in hrs in storing it in delta column
df['delta']=(df['Event Time'] - df['Install Time']).dt.total_seconds()/3600
```

```python
# finding the media source with biggest delta
media_source_with_max_delta= df.groupby('Media Source')['delta'].max().reset_index().round(2)
media_source_with_max_deltaa=media_source_with_max_delta.sort_values(by='delta', ascending=False)
media_source_with_max_deltaa
```

```python
plt.figure(figsize=(4, 3))
sns.barplot(data=media_source_with_max_deltaa, x='Media Source', y='delta', color='yellow')
plt.title('Media Source With Biggest Delta')
plt.xticks(rotation=45)
plt.xlabel('Media Source')
plt.ylabel('Delta')
plt.show()
```

```python
# calculating avg delta by Event Name
avg_delta_by_event_name=df.groupby('Event Name')['delta'].mean().reset_index().rename(columns={'delta':'avg_delta'}).round(2)
avg_delta_by_event_namee=avg_delta_by_event_name.sort_values(by='avg_delta', ascending=False)
```

```python
plt.figure(figsize=(4, 3))
sns.barplot(data=avg_delta_by_event_namee, x='Event Name', y='avg_delta', color='yellow')
plt.title('Avg Delta by Event Name')
plt.xticks(rotation=45)
plt.xlabel('Event Name')
plt.ylabel('Avg_Delta')
plt.show()
```
