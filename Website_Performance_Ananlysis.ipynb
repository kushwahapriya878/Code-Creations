```python
# Importing laibraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset and ignoring the first row as in this dataset first row has some error
df= pd.read_csv('data-export.csv', skiprows=1)
df
```

```python
# Converting the datetime in proper datetime format
df['Date']=pd.to_datetime(df['Date + hour (YYYYMMDDHH)'], format='%Y%m%d%H')
```

```python
df.drop(columns='Date + hour (YYYYMMDDHH)', inplace=True)
```

```python
# Renaming col name
df.rename(columns={'Session primary channel group (Default channel group)':'Channel_group'}, inplace=True)
```

```python
# group data by date and sum Users and sessions
grp_data=df.groupby('Date').agg({'Users':'sum', 'Sessions':'sum'})
grp_data.head(3)
```

```python
plt.figure(figsize=(15, 6))
plt.plot(grp_data.index, grp_data['Users'], label='Users')
plt.plot(grp_data.index, grp_data['Sessions'], label='Sessions')
plt.title('Total Users and Sessions Over Time')
plt.xlabel('Date & Hour')
plt.ylabel('Count')
plt.grid(True)
plt.legend()
plt.show()
```

# From the graph both total users and sessions following the same trend, when users are increasing sessions is also increasing. cycles of users and sessions are fluctuating, peak might correspond to some marketing activity.

# Let's analyse the User Engagment .We will look into metrics like 	Average engagement time per session, Engaged sessions per user, Events per session to evalute how engaged users are when they visit site.

```python
engagement_metrics=df.groupby('Date').agg({
    'Average engagement time per session':'mean',
    'Engaged sessions per user':'mean',
    'Events per session':'mean',
    'Engagement rate':'mean'

})

# ploting engagemnet_metrics
fig, ax =plt.subplots(4, 1, figsize=(15, 20), sharex=True)

ax[0].plot(engagement_metrics.index, engagement_metrics['Average engagement time per session'], label='Avg Engagemnet Time', color='blue')
ax[0].set_title('Average engagement time per session')
ax[0].set_ylabel('Seconds')

ax[1].plot(engagement_metrics.index, engagement_metrics['Engaged sessions per user'], label='Engaged sessions/User', color='orange')
ax[1].set_title('Engaged sessions per user')
ax[1].set_ylabel('Ratio')

ax[2].plot(engagement_metrics.index, engagement_metrics['Events per session'], label='Events per session', color='purple')
ax[2].set_title('Events per session')
ax[2].set_ylabel('Count')

ax[3].plot(engagement_metrics.index, engagement_metrics['Engagement rate'], label='Engagement rate', color='green')
ax[3].set_title('Engagement rate')
ax[3].set_ylabel('Date & Hour')

for a in ax:
  a.legend()
  a.grid(True)

plt.tight_layout()
plt.show()
```

**Average engagement time per session:** As per above graph Avg. engagement time incresed when users found more relatable content.

**Engaged sessions per user:** This ratio is fluctuates slightly  but generally shows a good portion of session per user engaged. Peaks shows when user found more ralated content they spend more time.

**Events per session:**There are some variation in graph . Peak shows when when user found more interactive content then they used to visit.

**Engagement rate:** The engagement rate over time shows that many sessions are considered engaged out of total.But there are some ups and down that  how different content resonates with users.

**Engaged sessions per user vs Engagement rate:** Both Engaged sessions per user and Engagement rate highly corelated

# Now we will analyse channel performance that how different marketing channel contribute to traffic and engagement

```python
# group data by channel
channel_performance=df.groupby('Channel_group').agg({
    'Users':'sum',
    'Sessions':'sum',
    'Engaged sessions':'sum',
    'Events per session':'mean',
    'Engagement rate':'mean'

})

# normalizing Engagement rate and Events per session for comperision
channel_performance['Normalized_Engagemnet_rate']=channel_performance['Engagement rate']/channel_performance['Engagement rate'].max()
channel_performance['Normalized_Events_per_session']=channel_performance['Events per session']/channel_performance['Events per session'].max()

#plotting the channel performance metrics

fig, ax=plt.subplots(3,1, figsize=(10,12))

#users by channel and sessions
ax[0].bar(channel_performance.index, channel_performance['Users'], label='Users', alpha=0.8, color='blue')
ax[0].bar(channel_performance.index, channel_performance['Sessions'], label='Sessions', alpha=0.8, color='orange')
ax[0].set_title('Users and Sessions by Channel')
ax[0].set_ylabel('count')
ax[0].legend()

#Normalized Engagement rate by channel
ax[1].bar(channel_performance.index, channel_performance['Normalized_Engagemnet_rate'], color='purple')
ax[1].set_title('Normalized Engagement Rate by Channel')
ax[1].set_ylabel('Normalized_rate')

#Normalized Event per sessions
ax[2].bar(channel_performance.index, channel_performance['Normalized_Events_per_session'], color='blue')
ax[2].set_title('Normalized Event/Sessions by Channel')
ax[2].set_ylabel('Normalized_count')

plt.tight_layout()
plt.show()
```

***The data illustrates significane variation in across differenet channel performance. above graph shows strength and weekness of each channel.Organic and referral both users are same while organicnically users engaging more than the referral.***
