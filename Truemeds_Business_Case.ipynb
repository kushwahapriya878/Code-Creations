# Applying Central Limit Theorem to Calculate Mean Event Revenue

```python
#importing laibraries
from scipy import stats
```

```python
df_c['Event Revenue'].mean()
```

```python
plt.figure(figsize=(10, 4))
sns.histplot(df_c['Event Revenue'])
plt.title('Distribution of Event Revenue')
plt.xlabel("Event Revenue")
plt.show()
```

```python
sample_means=[]
sample_size=30

for i in range(1000):
  sample=df_c['Event Revenue'].sample(sample_size)
  sample_means.append(sample.mean())
```

```python
plt.figure(figsize=(10, 4))
sns.histplot(sample_means, kde=True)
plt.title('Distribution of Sample Means')
plt.xlabel('Sample Mean')
plt.show()
```

```python
mean_of_sample_means= np.mean(sample_means)
print(f'mean_of_sample_mean:{mean_of_sample_means}')

std_of_sample_mean = np.std(sample_means)
print(f'std_of_sample_mean: {std_of_sample_mean}')

std_error_of_sample_mean= std_of_sample_mean/np.sqrt(len(sample_means))
print(f'std_error_of_sample_mean:{std_error_of_sample_mean}')

confidence_level= 0.95

confidence_interval = stats.t.interval(confidence_level, len(sample_means)-1, loc=mean_of_sample_means, scale=std_error_of_sample_mean)
print(f'95% confidence interval:{confidence_interval}')
```

# Insights:

The normal distribution of sample means which is 914 confirms that CLT's sampleing distribution of sample mean will approx follow the normal distring as we increase the sample size sample mean will get closure to population mean

At 95% confidence interval mean of sample means and population mean is falling in CI_range (901.38, 912.88)

WITH OrdersWithQuarters AS
( SELECT
cd.CustomerID,
cd.CreatedOn AS CustomerCreatedOn,
od.OrderID, od.CreatedOn AS OrderCreatedOn,
os.OrderStatusID,
DATEPART(QUARTER, od.CreatedOn) AS OrderQuarter,
DATEPART(YEAR, od.CreatedOn) AS OrderYear
FROM CxDetails cd
INNER JOIN
OrderDetails od
ON cd.CustomerID = od.CustomerID
INNER JOIN
OrderStatus os
ON od.OrderID = os.OrderID
WHERE os.OrderStatusID = 'Completed') ,
RecentOrders AS (
SELECT
CustomerID,
OrderQuarter,
OrderYear,
DENSE_RANK() OVER (ORDER BY OrderYear DESC, OrderQuarter DESC) AS
QuarterRank
FROM
OrdersWithQuarters ) ,
CurrentAndPreviousQuarters AS (
SELECT
CurrentQuarter.OrderYear,
CurrentQuarter.OrderQuarter,
COUNT(DISTINCT CurrentQuarter.CustomerID) AS CurrentQuarterCustomers,
COUNT(DISTINCT PreviousQuarter.CustomerID) AS PreviousQuarterCustomers
FROM RecentOrders CurrentQuarter
LEFT JOIN RecentOrders PreviousQuarter
ON CurrentQuarter.CustomerID = PreviousQuarter.CustomerID
AND CurrentQuarter.QuarterRank = PreviousQuarter.QuarterRank + 1
WHERE CurrentQuarter.QuarterRank <= 4 GROUP BY CurrentQuarter.OrderYear,
CurrentQuarter.OrderQuarter )
SELECT
OrderYear,
OrderQuarter,
CurrentQuarterCustomers,
PreviousQuarterCustomers,
(CurrentQuarterCustomers * 1.0 / NULLIF(PreviousQuarterCustomers, 0)) * 100 AS
RetentionRate
FROM CurrentAndPreviousQuarters
ORDER BY OrderYear, OrderQuarter;
